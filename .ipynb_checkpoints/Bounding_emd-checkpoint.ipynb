### Bounding

```python
unique_res = df['Res'].unique()
unique_scale = df['Scale_Param'].unique()
new_df = pd.DataFrame()
```

```python
for i, res_option in enumerate(unique_res):
    for j, scale_option in enumerate(unique_scale):
        
        filtered_df = df[(df['Res'] == res_option) & (df['Scale_Param'] == scale_option)]

        slope, intercept = np.polyfit(noises, filtered_df['Distances_Noised'], 1)
        
        new_row = {
            'Res' : res_option,
            'Scale_Param' : scale_option,
            'Slope' : slope,
            'Intercept' : intercept,
            'Distance': filtered_df['Distances_Noised'].values
        }

        new_df = new_df._append(new_row, ignore_index=True)
```

```python
def interactive_distance_fixed_res(index):
    curr_df = new_df[new_df['Res'] == unique_res[index]]
    plt.plot(curr_df['Scale_Param'], curr_df['Slope'])  # X first, Y second
    plt.xlabel('Scale')
    plt.ylabel('Slope')
```

```python
index_slider = widgets.IntSlider(
    value=0,
    min=0,
    max=len(unique_res) - 1,
    step=1,
    description='Resolution:',
    continuous_update=False,
    readout=True
)

# Create an interactive plot with a slider for noise_param
interactive_plot = interactive(interactive_distance_fixed_res, index=index_slider)
output = interactive_plot.children[-1]
output.layout.height = '440px'
interactive_plot
```

```python
def interactive_distance_fixed_scale(index):
    curr_df = new_df[new_df['Scale_Param'] == unique_scale[index]]
    plt.plot(curr_df['Res'], curr_df['Slope'])  # X first, Y second
    plt.xlabel('Res')
    plt.ylabel('Slope')
```

```python
index_slider = widgets.IntSlider(
    value=0,
    min=0,
    max=len(unique_scale) - 1,
    step=1,
    description='Scale:',
    continuous_update=False,
    readout=True
)

# Create an interactive plot with a slider for noise_param
interactive_plot = interactive(interactive_distance_fixed_scale, index=index_slider)
output = interactive_plot.children[-1]
output.layout.height = '440px'
interactive_plot
```

### Approximating the distances

```python
approximated_distances_df = pd.DataFrame()
```

```python
for scale in unique_scale:
    for res in unique_res:
        for noise in noises:
            relevant_row = df[(df['Scale_Param'] == scale) & 
            (df['Res'] == res) &
            (df['Noise_Param'] == noise)]
            relevant_row_slopes = new_df[(new_df['Res'] == res) & (new_df['Scale_Param'] == scale)]
            slope = relevant_row_slopes['Slope'].values[0]
            approximated_distance = 0.3 * scale + noise * slope

            classic_distance = relevant_row['Distances_Classic'].values[0]
            noised_distance = relevant_row['Distances_Noised'].values[0]
            
            new_row = {
                'Scale' : scale,
                'Res' : res,
                'Noise' : noise,
                'Distance without noise' : classic_distance,
                'Distance with Noise' : noised_distance,
                'Distance with noise bound' : approximated_distance,
                'Ratio' : noised_distance / approximated_distance
            }
            
            approximated_distances_df = approximated_distances_df._append(new_row, ignore_index=True)
```

```python
len(approximated_distances_df.index)
```

```python
approximated_distances_df.head()
```

```python
plt.plot(approximated_distances_df['Ratio'])
```

Here, we notice that using the linear approximation of :

$W_{1} (p',q') = \delta * S + f(S,R) * N$  

Where:  
S stands for Scale  
R stands for Resolution  
N stands for Noise  
$\delta$ is the distance between the means of the gaussians (p,q)

```python
approximated_distances_df['Ratio'].mean()
```

```python
values = approximated_distances_df['Ratio'].values
lower = np.percentile(values, 2.5)
upper = np.percentile(values, 97.5)

print(f"95% of the values lie between {lower} and {upper}")
```

```python
plt.figure(figsize=(10,10))
working_res_index = 2
for i in range(10):
    subset = approximated_distances_df[(approximated_distances_df['Scale'] == unique_scale[i]) & 
                                    (approximated_distances_df['Res'] == unique_res[working_res_index])]
    plt.plot(subset['Noise'], subset['Ratio'], label = f'Scale={unique_scale[i]}')
plt.title(f'Distance ratio given different scale values, and res={unique_res[working_res_index]}')
plt.xscale('log')
plt.xlabel('Noise')
plt.ylabel('Ratio')
plt.legend()
plt.show()
```

We see a weird phenomena in which there is a drop in ratios for the middle of the graph, for every single scale. This makes sense because scale simply multiplies all the values.

```python
plt.figure(figsize=(10,10))
working_scale_index = 4
for i in range(10):
    subset = approximated_distances_df[
    (approximated_distances_df['Scale'] == unique_scale[working_scale_index]) & 
                                    (approximated_distances_df['Res'] == unique_res[i])]
    plt.plot(subset['Noise'], subset['Ratio'], label = f'Resolution={unique_res[i]}')
plt.title(f'Distance ratio given different Res values, and scale={unique_scale[working_scale_index]}')
plt.xscale('log')
plt.legend()
plt.show()
```
