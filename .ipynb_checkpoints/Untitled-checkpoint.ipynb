## Exploring noise with Fourier

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fftn, ifftn
import scipy.interpolate 
import cv2
import time
import itertools
import ot
import glob
import os
from utils.utils import *
```

```python
def f_metric(mu, nu, s=1, p=2, alpha=0):
    """
    Calculates the f_{s,p}^{(alpha)} metric between two measures mu and nu.

    Args:
        mu: A NumPy array representing the first measure.
        nu: A NumPy array representing the second measure.
        s: The smoothness parameter (default: 1).
        p: The power in the distance function (default: 2).
        alpha: Weighting for low-frequency components (default: 0).

    Returns:
        A float representing the  f_{s,p}^{(alpha)} metric between mu and nu.
    """

    # Check if measures have the same shape
    if mu.shape != nu.shape:
        raise ValueError("Measures must have the same shape.")

    # Compute Fourier transforms 
    mu_fft = np.fft.fftn(mu)
    nu_fft = np.fft.fftn(nu)

    # Calculate frequencies 
    freqs = np.fft.fftfreq(mu.shape[0], d=mu.ndim)  
    grid = np.meshgrid(*freqs, indexing='ij')  
    k_mag = np.sqrt(np.sum(np.square(grid), axis=0))  

    # Compute the weighted difference
    diff = np.abs(mu_fft - nu_fft)**p
    weights = np.where(k_mag == 0, 0, k_mag**(-s * p - alpha))
    weighted_diff = weights * diff

    # Integrate (using a Riemann sum approximation)
    integral = np.sum(weighted_diff) * np.prod(mu.shape)**(-1 / mu.ndim) 

    # Calculate the metric
    metric = (integral)**(1/p)

    return metric
```

```python
def resample_image(image, target_resolution):
    """
    Resamples an image to a given target resolution using interpolation.

    Args:
        image: The image to resample (as a NumPy array).
        target_resolution: The desired resolution (height, width).

    Returns:
        The resampled image as a NumPy array.
    """
    # Choose an appropriate interpolation method (e.g., 'linear', 'cubic')
    interpolation_method = 'linear' 

    # Calculate scaling factors
    scale_y = target_resolution[0] / image.shape[0]
    scale_x = target_resolution[1] / image.shape[1]

    # Create grid for interpolation
    grid_y, grid_x = np.mgrid[0:target_resolution[0], 0:target_resolution[1]]

    # Perform interpolation
    resampled_image = scipy.interpolate.interp2d(
        np.arange(image.shape[1]), 
        np.arange(image.shape[0]), 
        image, 
        kind=interpolation_method
    )(scale_x * grid_x, scale_y * grid_y)

    return resampled_image
```

```python
def calculate_pfm(image1_path, image2_path, target_resolution=None, s=1, p=2, alpha=0):
    """
    Calculates the Periodic Fourier Metric (PFM) between two images.

    Args:
        image1_path: The path to the first image.
        image2_path: The path to the second image.
        target_resolution:  If provided (tuple of height, width), images are 
                            resampled to match this resolution.
        s: Smoothness parameter (default: 1).
        p: Power in the distance function (default: 2).
        alpha: Weighting for low-frequency components (default: 0).

    Returns:
        The Periodic Fourier Metric between the two images.
    """

    # Load images
    image1 = cv2.imread(image1_path)
    image2 = cv2.imread(image2_path)

    # Resample if a target resolution is provided
    if target_resolution is not None:
        image1 = resample_image(image1, target_resolution)
        image2 = resample_image(image2, target_resolution)

    # Ensure images now have the same shape
    if image1.shape != image2.shape:
        raise ValueError("Images must have the same dimensions after resampling.")

    # Calculate the PFM (using our previously defined f_metric function)
    pfm = f_metric(image1, image2, s, p, alpha) 

    return pfm
```

```python
# Parameters for the sine wave and noise
n_points = 500
x = np.linspace(0, 4 * np.pi, n_points)
original_signal = np.sin(x)
noise_variance = 0.5
noise = np.random.normal(0, np.sqrt(noise_variance), n_points)
noisy_signal = original_signal + noise
```

```python
# Compute the Fourier Transform of both signals
original_signal_fft = np.fft.fft(original_signal)
noisy_signal_fft = np.fft.fft(noisy_signal)
frequencies = np.fft.fftfreq(n_points, d=(x[1]-x[0])/(2*np.pi))
```

```python
# Plotting
fig, ax = plt.subplots(2, 2, figsize=(10, 7))

# Time domain plots
ax[0, 0].plot(x, original_signal, label="Original Signal")
ax[0, 0].set_title("Original Signal in Time Domain")
ax[0, 1].plot(x, noisy_signal, label="Noisy Signal", color="orange")
ax[0, 1].set_title("Noisy Signal in Time Domain")

# Frequency domain plots
ax[1, 0].plot(frequencies, np.abs(original_signal_fft), label="Original FFT")
ax[1, 0].set_title("FFT of Original Signal")
ax[1, 1].plot(frequencies, np.abs(noisy_signal_fft), label="Noisy FFT", color="orange")
ax[1, 1].set_title("FFT of Noisy Signal")

for a in ax.flat:
    a.legend()
    a.grid(True)

plt.tight_layout()
plt.show()
```

```python
def generate_simple_image(size=256, shape='circle'):
    """
    Generates a simple 2D image with a specified shape.
    Currently supports a circle or a square in the center of the image.
    """
    image = np.zeros((size, size))
    if shape == 'circle':
        center = (size // 2, size // 2)
        radius = size // 4
        for x in range(size):
            for y in range(size):
                if (x - center[0])**2 + (y - center[1])**2 <= radius**2:
                    image[x, y] = 1
    elif shape == 'square':
        start = size // 4
        end = 3 * size // 4
        image[start:end, start:end] = 1
    return image

# Generate a simple image and add white Gaussian noise
size = 256
original_image = generate_simple_image(size, 'circle')
noise_variance = 0.1
noise = np.random.normal(0, np.sqrt(noise_variance), (size, size))
noisy_image = original_image + noise

# Compute the Fourier Transform of both images
original_image_fft = np.fft.fftshift(np.fft.fft2(original_image))
noisy_image_fft = np.fft.fftshift(np.fft.fft2(noisy_image))

# Visualizing the images and their Fourier Transforms
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

# Original and noisy images
ax[0, 0].imshow(original_image, cmap='gray')
ax[0, 0].set_title("Original Image")
ax[0, 1].imshow(noisy_image, cmap='gray')
ax[0, 1].set_title("Noisy Image")

# Fourier Transforms
ax[1, 0].imshow(np.log(np.abs(original_image_fft) + 1), cmap='gray')
ax[1, 0].set_title("FFT of Original Image")
ax[1, 1].imshow(np.log(np.abs(noisy_image_fft) + 1), cmap='gray')
ax[1, 1].set_title("FFT of Noisy Image")

for a in ax.flat:
    a.axis('off')

plt.tight_layout()
plt.show()
```

### Calculate the F metric for a single sample

```python
p = 2
s = 2
T = 32
d = 2
```

```python
def calculate_2d_frequencies(n_rows, n_cols):
    # Compute frequency values for each axis
    fy = np.fft.fftfreq(n_rows).reshape(-1, 1)  # Frequencies in y-direction
    fx = np.fft.fftfreq(n_cols)  # Frequencies in x-direction
    
    # Create a meshgrid of frequencies: combines fx and fy to form a grid of vectors
    fx, fy = np.meshgrid(fx, fy)
    
    # Calculate the magnitude of the frequency vectors
    # This represents |k|, the distance from the origin to each point in frequency space
    freq_magnitude = np.sqrt(fx**2 + fy**2)
    
    return freq_magnitude
```

```python
diff = np.abs(im1_ft - im2_ft) ** p
freqs = calculate_2d_frequencies(*im1_ft.shape)

freqs[freqs == 0] = 1e-5

weighted_diff = diff / (freqs ** (s*p))

pfm = ((np.sum(weighted_diff)) ** (1/p)) / (T**d)
```
